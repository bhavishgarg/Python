```python
# Maintaining a tail:
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def printLL(head):
    while head is not None:
        print(str(head.data) + "->", end = "")       # We're also printing -> so we need to convert head.data to string
        head = head.next
    print("None")
    return
        
def takeInput():                                      # Now, this logic takes very less time complared to O(N^2)
    arr = list(map(int,input().split()))
    head = None
    tail = None
    for ele in arr:
        if ele == -1:
            break
        
        newNode = Node(ele)

        if head is None:
            head = newNode
            tail = newNode
            
        else:                            # tail will move one step ahead after each iteration
            tail.next = newNode          # connecting the previous last node to the new last node
            tail = newNode               # tail will start pointing to the new last node
            
    return head 

def mid_point(head):
    slow=head
    fast=head  # Note the AND here
    while fast.next!=None and fast.next.next!=None:     # first condition for odd no: of elements,second cond. for even no:
        slow=slow.next                                  # Slow will increase at speed x
        fast=fast.next.next                             # Fast will increase at speed 2x 
    return slow

def mergeLL(head1, head2):
    
    if head1 is None:
        return head2
    if head2 is None:
        return head1

    # get the correct first node and increment the pointer
    if head1.data < head2.data:
        head = head1
        head1 = head1.next
    else:
        head = head2
        head2 = head2.next
    current = head
    
    # while both of the lists have data
    while head1 and head2:
        # merge the lower value node
        if head1.data < head2.data:
            current.next = head1
            head1 = head1.next
        else:
            current.next = head2
            head2 = head2.next
        # and update the pointer
        current = current.next
        
    # if there are nodes remaining in one of the lists,
    # append it to the result
    if head1:
        current.next = head1
    elif head2:
        current.next = head2
    return head 

def mergesort(head):
 
    # base case â€” length 0 or 1
    if head is None or head.next is None:
        return head
 
    # split `head` into `a` and `b` sublists
    mid=mid_point(head)
    head1=head
    head2=mid.next
    mid.next=None
    
    # recursively sort the sublists
    head1 = mergesort(head1)
    head2= mergesort(head2)
    
    # answer = merge the two sorted lists
    return mergeLL(head1,head2)          


head = takeInput()
sorted_head=mergesort(head)
printLL(sorted_head)
```
